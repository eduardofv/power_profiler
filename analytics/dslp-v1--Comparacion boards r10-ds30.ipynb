```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fn1 = "../data/dslp--wemosc3mini_r10_ds30-T1.csv"
fn2 = "../data/dslp--wemosd1_r10_ds30-T1.csv"
fn3 = "../data/dslp--nodemcu_r10_ds30-T2.csv"
fn4 = "../data/dslp--saola_r20_ds30-T1.csv"
```

```python
df1 = pd.read_csv(fn1, parse_dates=["read_at", "timestamp"])
df1.columns = ['index','read_at', 'millis', "dt", "v", "a", "w", "v_sh"]
df2 = pd.read_csv(fn2, parse_dates=["read_at", "timestamp"])
df2.columns = ['index','read_at', 'millis', "dt", "v", "a", "w", "v_sh"]
df3 = pd.read_csv(fn3, parse_dates=["read_at", "timestamp"])
df3.columns = ['index','read_at', 'millis', "dt", "v", "a", "w", "v_sh"]
df4 = pd.read_csv(fn4, parse_dates=["read_at", "timestamp"])
df4.columns = ['index','read_at', 'millis', "dt", "v", "a", "w", "v_sh"]

plt.plot(df1.a)
plt.plot(df2.a)
plt.plot(df3.a)
plt.plot(df4.a)
```

```python
plt.plot(df1.a[df1.a < 0.0125])
plt.plot(df2.a[df2.a < 0.0125])
plt.plot(df3.a[df3.a < 0.0125])
plt.plot(df4.a[df4.a < 0.0125])
```
